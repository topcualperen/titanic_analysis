```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
df_clean = df[['Pclass', 'Survived', 'Age', 'Sex']].copy()
```

```python
# Grafik 1: Sınıflara göre hayatta kalma oranı (Bar Plot)
survival_by_class = df_clean.groupby('Pclass')['Survived'].mean()
```

```python
plt.figure(figsize=(8, 5))
survival_by_class.plot(kind='bar', color=['#FF6B6B', '#4ECDC4', '#45B7D1'])
plt.title('Sınıflara Göre Hayatta Kalma Oranı')
plt.xlabel('Yolcu Sınıfı')
plt.ylabel('Hayatta Kalma Oranı')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Çıkarım: 1. sınıf yolcuların hayatta kalma oranı 3. sınıfa göre 2 kat daha yüksek.
# Sosyo-ekonomik durum, hayatta kalmada kritik rol oynamış.
```

```python
# Grafik 2: Yaş dağılımı (Histogram)
plt.figure(figsize=(8, 5))
plt.hist(df_clean['Age'].dropna(), bins=30, color='#95E1D3', edgecolor='black')
plt.title('Yolcuların Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Kişi Sayısı')
plt.tight_layout()
plt.show()
```
